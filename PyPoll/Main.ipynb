# PyPoll

```python
import os
import csv
```

```python
# reading in the csv
csvpath = os.path.join("election_data.csv")

# appending the data to a variable after it is read in
pypoll = []

with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter = ",")

    for x in csvreader:
        # where the data is appended
        pypoll.append(x)
```

```python
pypoll
```

```python
# getting the total number of votes
total_votes = len(pypoll)
```

```python
total_votes
```

```python
# creating a set
candidates = set()

# iterating over the data
for i in range(1, len(pypoll)):
    candidate = pypoll[i][2]
    
    # appending the names to the set
    candidates.add(candidate)
```

```python
# getting the unique names of each canadidate
candidates
```

```python
#getting percentage of votes each candidate won


vote_counts = {}
for i in range(1,len(pypoll)):
    # appending candidate name from the 3rd column 
    candidate = pypoll[i][2]
    
    # if statement to hold each vote respectivley
    if candidate not in vote_counts:
        vote_counts[candidate] = 0
    
    # counting up 1 for each instance of a vote for a candidate
    vote_counts[candidate] += 1

# get total number of votes
total_votes = sum(vote_counts.values())

# Calculate the percentage of votes each candidate won
vote_percentages = {candidate: (votes / total_votes) * 100 for candidate, votes in vote_counts.items()}
```

```python
vote_percentages
```

```python
vote_counts = {}

# Iterate over the data, starting from the second row (to skip the header)
for i in range(1, len(pypoll)):
    # The candidate name is the third item in each inner list
    candidate = pypoll[i][2]
    
    # If this candidate is not already in the dictionary, add them with a vote count of 0
    if candidate not in vote_counts:
        vote_counts[candidate] = 0
    
    vote_counts[candidate] += 1
```

```python
vote_counts
```

```python
# combining the two results into one for exporting to txt file
combined_data = {}
for candidate in vote_percentages:
    combined_data[candidate] = {"Percentage of votes": vote_percentages[candidate], "Total votes": vote_counts[candidate]}
```

```python
combined_data
```

```python
# exporting to txt file
with open('results_poll.txt', 'w') as file:
    for candidate, data in combined_data.items():
        file.write(f"{candidate}\n")
        for key, value in data.items():
            file.write(f"  {key}: {value}\n")
```
